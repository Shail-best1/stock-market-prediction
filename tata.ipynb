```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tatadata.csv')
df.head()
```

```python
array = df['Open'].to_numpy()
array = array[1:]
series = pd.Series(array)
df['Next Day'] = series
```

```python
df.head(5)
```

```python
df = df[df['Volume']>0]
df.head()
```

```python
df = df.drop(columns=['Adj Close','Date'])
df.sample(5)
```

```python
df['Next day %'] = ((df['Next Day'] - df['Open'])/df['Open'])*100
df['High %'] = ((df['High'] - df['Open'])/df['Open'])*100
df['Low %'] = ((df['Low'] - df['Open'])/df['Open'])*100
df['Close %'] = ((df['Close'] - df['Open'])/df['Open'])*100
df=df.drop(columns=['High','Low','Close'])
df.head(5)
```

```python
df=df.drop(columns=['Open','Next Day'])
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df
```
